```python
import numpy as np, pandas as pd,matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score , accuracy_score
import xgboost as xg
```

```python
df = pd.read_csv("/mnt/mydata/python_projects/python_projects/iris/data/Iris.csv")
df = df.drop_duplicates()
df.dropna()
```

```python
df['Species'].value_counts()
```

```python
species = {'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2}
df['Species'] = df['Species'].map(species)
```

```python
df
```

```python
X = pd.DataFrame({'SepalLengthCm':df['SepalLengthCm'],'SepalWidthCm':df['SepalWidthCm'],
                  'PetalLengthCm': df['PetalLengthCm'],'PetalWidthCm':df['PetalWidthCm']})
Y = pd.DataFrame({'Species':df['Species']})
x_train,x_test,y_train,y_test = train_test_split(X,Y,random_state=42,test_size=0.2)

model = xg.XGBClassifier(
    n_estimators=100,
    learning_rate=0.5,
    max_depth=10,
    tree_method='exact',
    objective='multi:softmax',
    random_state=42,
    num_class=3
)

# Fix: Convert y_train to 1D array
model.fit(x_train, y_train.values.ravel(), verbose=True)

# Fix: For prediction comparison
y_predict = model.predict(x_test)
result = pd.DataFrame({'Actual':y_test.values.ravel(),'Predict':y_predict})
result
```

```python
score = f1_score(y_test, y_predict,average='weighted') # <--- THIS IS THE POTENTIAL ISSUE
accuracy = accuracy_score(y_test, y_predict)
print(f'f1 score {score}\naccuracy score:{accuracy:.4f}')
```
