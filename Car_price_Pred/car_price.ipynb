```python
import numpy as np, pandas as pd,matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/mnt/mydata/python_projects/python_projects/iris/data/car data.csv")
df = df.drop_duplicates()
df.dropna()
```

```python
df['Car_Name'].values
Car_name = {}
fromdataset = df['Car_Name'].values 
for i in range(len(fromdataset)):
    if  fromdataset[i] not in Car_name:
        Car_name[fromdataset[i]] =i
Car_name
```

```python
df['Car_Name'] = df['Car_Name'].map(Car_name)
df
```

```python
Fuel = {'Diesel':0, 'Petrol':1}
df['Fuel_Type'] = df['Fuel_Type'].map(Fuel)
df
```

```python
df['Selling_type'] = np.where(df["Selling_type"] == "Dealer",1,0)
df
```

```python
df['Transmission'] = np.where(df['Transmission'] == 'Manual',1,0)
```

```python
x = {"Car_Name":df['Car_Name'].values,'Year':df['Year'].values,'Present_Price':df['Present_Price'].values,'Driven_kms':df['Driven_kms'].values,'Fuel_Type':df['Fuel_Type'].values,'Selling_type':df['Selling_type'].values,'Transmission':df['Transmission'].values,'Owner':df['Owner'].values}
x = pd.DataFrame(x)
y = {"Selling_Price":df["Selling_Price"]}
y = pd.DataFrame(y)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x,y,test_size=0.2,random_state=42)
import xgboost as xg
from sklearn.metrics import r2_score, mean_absolute_error
```

```python
X_train
```

```python
model =xg.XGBRegressor( objective='reg:gamma',
        n_estimators=100,
        learning_rate=0.6,
        max_depth=3,
        tree_method='hist',
        min_child_weight=3,
        random_state=42
         )
model.fit(X_train, Y_train,verbose=True)
```

```python
y_pred = model.predict(X_test)
r2score = r2_score(Y_test,y_pred)
absoluteerror= mean_absolute_error(Y_test,y_pred)
print(f'r2_score:{r2score:.2f}\nabsolute error {absoluteerror:.4f}')
```

```python
import joblib
joblib.dump(model,"./models/XGBRegressor.lb")
```
